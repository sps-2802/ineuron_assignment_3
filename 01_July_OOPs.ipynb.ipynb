```python
# 1.What is the primary goal of Object-Oriented Programming (OOP)/
#     The primary goal of Object-Oriented Programming is to organise and structure code.
#     In Python, you can create classes to define objects with variables and functions associated with them.
class Vehicle:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model
    def start(self):
        print(f'{self.brand} {self.model} is starting...')
    def stop(self):
        print(f'{self.brand} {self.model} is stopping...')
class Car(Vehicle):
    car = Vehicle('kia','sonnet')#object creation
    car.start()
    car.stop()
```

```python
# 3. What is a class in Python?
# A class is a blueprint or template that defines the structure and behaviour of an object.
# class keyword is used to define the class followed by a classname in camelCase convention and followed by colon. 

class Car:#class creation
    def __init__(self,brand,model,year):
        self.brand = brand
        self.model = model
        self.year = year
    def display(self):
        print(f'The brand of the car is {self.brand} model is {self.model} and year is {self.year}')
brand = input('Enter the brand of the car:')
model = input('Enter the model of the car:')
year = input('Enter the year of the car:')
car = Car(brand,model,year)#object creation
car.display()
```

```python
# 4. What are attributes and methods in a class?
# Attributes:
#     Attributes are the varaibles or data members associated with the class.
#     It represents the state or characteristics of an object.
#     Each object created from the class will have its own set of attribute values. 
# Methods:
#     Methods are functions that are defined within a class that define the behaviour of the class.
#     Methods define the operations or actions that can be performed on the data of an object. 
#     Methods can perform computations, manipulate data, or interact with other objects.

class Addition:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def addition(self,a,b): 
        return self.a+self.b

a = int(input('Enter the value of a:'))
b = int(input('Enter the value of b:'))
add = Addition(a,b)
res = add.addition(a,b)
print('The sum of two numbers are',res)
```

```python
# 5.What is the difference between class variables and instance variables in Python?
# Class variables and instance variables are both types of variables that can be defined within a class.
# class variables:
#     Class variables are defined within a class but outside any methods.
#     Class variables are allocated memory only once when the class is defined. 
#     When a class variable is modified, the change is reflected in all instances of the class.
# Instance variables:
#     Instance variables are defined within methods or the __init__ constructor method of a class.
#     Instance variables are accessed and modified through the individual instances of the class.
#     Instance variables are allocated memory separately for each instance of the class.

class Employee:
    company = 'ABC corporation'#class variable
    def __init__(self,name,salary):
        self.name = name #instance variable
        self.salary = salary# instance variable
    def display(self):
        print(f'name is {self.name} and salary is {self.salary}')
name = input('Enter the name:')
salary = int(input('Enter the salary'))
emp = Employee(name,salary)
emp.display()
print('Company Name:',emp.company)#accessing the class variables
    
```

```python
# 8. For a ticket booking system, you have to design the "Ticket" class with OOP
# principles in mind. The “Ticket” class should have the following attributes:
# a. ticket_id: Represents the unique identifier for the ticket.
# b. event_name: Represents the name of the event.
# c. event_date: Represents the date of the event.
# d. venue: Represents the venue of the event.
# e. seat_number: Represents the seat number associated with the ticket.
# f. price: Represents the price of the ticket.
# g. is_reserved: Represents the reservation status of the ticket.
# The class also includes the following methods:
# a. reserve_ticket(self): Marks the ticket as reserved if it is not already reserved.
# b. cancel_reservation(self): Cancels the reservation of the ticket if it is already
# reserved.
# c. display_ticket_info(self): Displays the information about the ticket, including its
# attributes and reservation status.
class Ticket:
    total_tickets = 100
    def __init__(self,ticket_id,event_name,event_date,venue,seat_number,price,is_reserved):
        self.ticket_id =ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = is_reserved
    def reserve_ticket(self):
        if self.is_reserved:
            self.is_reserved = 'reserved'
            print(f'Ticket {self.ticket_id} is successfull')
            if self.total_tickets <=100:
                self.total_tickets -= 1
            print(f'Total tickets available {self.total_tickets}')
            
        else:
            print(f'Ticket is reserved/filled')
    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = 'cancelled/not booked'
            print(f'Ticket {self.ticket_id} is cancelled and amount will be refunded within 3 to 5 working days')
            if self.total_tickets <100:
                self.total_tickets +=1
                print(f'Total tickets available {self.total_tickets}')
            else:
                self.total_tickets+=0
                print(f'Total tickets available {self.total_tickets}')
    def display_ticket_info(self):
        print(f'''
        --------------Ticket----------------
        Ticket ID           : {self.ticket_id} 
        Event Name          : {self.event_name}
        Event Date          : {self.event_date}
        Venue               : {self.venue}
        Seat Number         : {self.seat_number}
        Price               : {self.price}
        Reseved/Not Reserved:  {self.is_reserved}
      --------------------------------------''')
class Main(Ticket):
    while True:
        print(''' Ticket Reservation System
                  1. Reserve Ticket
                  2. Cancel Ticket
                  3. display Ticket
                  4. exit
        ''')
        choice = int(input('Enter your choice:'))
        match choice:
            case 1:
                ticket_id = input('enter the ticket id: ')
                event_name = input('enter the event name:')
                event_date = input('enter the event date:')
                venue = input('enter the venue:')
                seat_number = input('enter the seat number:')
                price = 1000
                is_reserved = 'reserved'
                main = Ticket(ticket_id,event_name,event_date,venue,seat_number,price,is_reserved)
                main.reserve_ticket()
            case 2:
                main.cancel_reservation()
            case 3:
                main.display_ticket_info()
            case default:
                break
        
```

```python
# 9.You are creating a shopping cart for an e-commerce website. Using OOP to model
# the "ShoppingCart" functionality the class should contain following attributes and
# methods:
# a. items: Represents the list of items in the shopping cart.
# The class also includes the following methods:

# a. add_item(self, item): Adds an item to the shopping cart by appending it to the
# list of items.
# b. remove_item(self, item): Removes an item from the shopping cart if it exists in
# the list.
# c. view_cart(self): Displays the items currently present in the shopping cart.
# d. clear_cart(self): Clears all items from the shopping cart by reassigning an
# empty list to the items attribute.


class ShoppingCart:
    items = []
    def add_item(self,item):
        if item not in self.items:
            self.items.append(item)
            print(f'{item} successfully added to the cart')
        else:
            print(f'{item} already in the cart')
    def remove_item(self,item):
        if item in self.items:
            self.items.remove(item)
            print(f'{item} successfully removed from the cart')
        else:
            print(f'{item} is not available in the cart')
    def view_cart(self):
        if self.items:
            print(f'Products available in the cart {self.items}')
        else:
            print('No products in the cart')
    def clear_cart(self):
        if self.items:
            self.items = []
            print('All items cleared in the cart..')
        else:
            print('No products in the cart')
shoppingcart = ShoppingCart()
while True:
    print('''\n------Menu-----\n
    1. Add Item to cart\n
    2. View Cart\n
    3. Remove Item from cart\n
    4. Clear cart\n
    5. Exit\n''')
    choice = int(input('Enter your choice:'))
    match choice:
        case 1:
            numberOfItems = int(input('\n\nEnter the Number of items to be added:'))
            for i in range(numberOfItems):
                item = input('\n\nEnter the item to be added to the list:')
                shoppingcart.add_item(item)
        case 2:
            shoppingcart.view_cart()
        case 3:
            item = input("\nEnter the item you want to remove: ")
            shoppingcart.remove_item(item)
        case 4:
            shoppingcart.clear_cart()
        case default:
            break
```
